These can also be seen my simply printing the dataframe with the describe function. In the box plots, data sets I, II, and III seem to have the same Inner Quartile ranges and maxes and mins in the X data while IV seems to have all data at 8 and an upper outlier. The Y values seem have about the same IQR for III and IV each with upper outliers, and the IQRs for I and II look similar but they have slightly different medians and a very different max, II also has a lower outlier where I does not. Using the line plots, it is hard to make comparisons and differentiate between data sets as they are very scattered and there is no uniform trends in the line plots. But by it can be seen that there was more variation in the line plots for I and II as they are more scattered and line plots for III and IV have a less chaotic assortment of lines in their graphs. Through these graphs, it can be reasoned that IV had the least even spread of data in its set, with most data falling at 8 but having an outlier at 19, and the others had similar even spreads with a defined IQR, min, max, and median in X. In Y, III and IV had similar IQRs with upper outliers and I and II also had similar IQRs but with different maxes and II had a lower outlier. Line plots showed much variation in I and II, and more constant values with upper outliers for III and IV.
